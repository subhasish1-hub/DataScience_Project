We first need to create a folium `Map` object, with an initial center location to be NASA Johnson Space Center at Houston, Texas.

We could use `folium.Circle` to add a highlighted circle area with a text label on a specific coordinate. For example,


```python
# Create a blue circle at NASA Johnson Space Center's coordinate with a popup label showing its name
circle = folium.Circle(nasa_coordinate, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('NASA Johnson Space Center'))
# Create a blue circle at NASA Johnson Space Center's coordinate with a icon showing its name
marker = folium.map.Marker(
    nasa_coordinate,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'NASA JSC',
        )
    )
site_map.add_child(circle)
site_map.add_child(marker)
```

```python
import folium

# Define the launch sites with their names and coordinates
launch_sites = {
    'CCAFS LC-40': {'latitude': 28.562302, 'longitude': -80.577356},
    'CCAFS SLC-40': {'latitude': 28.563197, 'longitude': -80.576820},
    'KSC LC-39A': {'latitude': 28.573255, 'longitude': -80.646895},
    'VAFB SLC-4E': {'latitude': 34.632834, 'longitude': -120.610745}
}

# Calculate the center of the map to include all locations
latitudes = [site['latitude'] for site in launch_sites.values()]
longitudes = [site['longitude'] for site in launch_sites.values()]
center_lat = sum(latitudes) / len(latitudes)
center_lon = sum(longitudes) / len(longitudes)

# Create a folium Map object, centered to show all locations
# *** KEY CHANGE HERE: Reduced zoom_start to 2 for a wider view ***
site_map = folium.Map(location=[center_lat, center_lon], zoom_start=5)

# Add a circle and a marker for each launch site
for site_name, coords in launch_sites.items():
    # Add a blue circle for each location
    folium.Circle(
        location=[coords['latitude'], coords['longitude']],
        radius=1000,  # Radius in meters
        color='#0000FF', # Blue color
        fill=True,
        fill_color='#0000FF',
        fill_opacity=0.2,
        tooltip=site_name
    ).add_to(site_map)

    # Add a custom marker with the site name as text
    folium.Marker(
        location=[coords['latitude'], coords['longitude']],
        icon=folium.DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12px; color:#d35400;"><b>{site_name}</b></div>',
        )
    ).add_to(site_map)

# To display the map directly in Jupyter, simply ensure the map object is the last line
site_map
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
import folium
from folium.plugins import MarkerCluster

# Assuming spacex_df is already loaded and contains 'Launch Site', 'Lat', 'Long', and 'Class' columns.
# For demonstration purposes, let's create a dummy spacex_df if it's not already defined:
# This part is just for testing and won't be needed if you already have spacex_df
import pandas as pd
data = {
    'Launch Site': ['KSC LC-39A', 'KSC LC-39A', 'KSC LC-39A', 'CCAFS SLC-40', 'CCAFS SLC-40',
                    'CCAFS SLC-40', 'CCAFS SLC-40', 'CCAFS SLC-40', 'CCAFS SLC-40', 'CCAFS SLC-40'],
    'Lat': [28.573255, 28.573255, 28.573255, 28.563197, 28.563197,
            28.563197, 28.563197, 28.563197, 28.563197, 28.563197],
    'Long': [-80.646895, -80.646895, -80.646895, -80.576820, -80.576820,
             -80.576820, -80.576820, -80.576820, -80.576820, -80.576820],
    'Class': [1, 1, 1, 1, 1, 1, 0, 0, 1, 0]
}
spacex_df = pd.DataFrame(data)


# 1. Create a MarkerCluster object
marker_cluster = MarkerCluster()

# 2. Create a new column in spacex_df dataframe called marker_color
#    to store the marker colors based on the 'Class' value

def get_marker_color(launch_class):
    if launch_class == 1:
        return 'green'
    else:
        return 'red'

spacex_df['marker_color'] = spacex_df['Class'].apply(get_marker_color)

# 3. For each launch result in spacex_df data frame, add a folium.Marker to marker_cluster

# First, let's redefine the launch_sites dictionary from Task 1,
# as we'll need the center_lat and center_lon to initialize the map.
launch_sites = {
    'CCAFS LC-40': {'latitude': 28.563197, 'longitude': -80.576820},
    'CCAFS SLC-40': {'latitude': 28.563197, 'longitude': -80.576820}, # Duplicated as in image, assume for now
    'KSC LC-39A': {'latitude': 28.573255, 'longitude': -80.646895},
    'VAFB SLC-4E': {'latitude': 34.632834, 'longitude': -120.610745}
}

# Calculate the center of the map (re-using logic from Task 1)
latitudes = [site['latitude'] for site in launch_sites.values()]
longitudes = [site['longitude'] for site in launch_sites.values()]
center_lat = sum(latitudes) / len(latitudes)
center_lon = sum(longitudes) / len(longitudes)

# Create a Folium map object, centered to show all Locations
site_map = folium.Map(location=[center_lat, center_lon], zoom_start=5)

# Add the MarkerCluster to the map
site_map.add_child(marker_cluster)

for index, record in spacex_df.iterrows():
    # Create a Folium marker with the appropriate color and add it to the marker cluster
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        icon=folium.Icon(color='white', icon_color=record['marker_color']),
        popup=f"Launch Site: {record['Launch Site']}<br>Class: {record['Class']}"
    )
    marker_cluster.add_child(marker)

# To display the map directly in Jupyter, ensure the map object is the last line
# of the cell.
site_map
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
